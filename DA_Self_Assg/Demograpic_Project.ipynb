```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xlsxwriter

# Load and clean your data
file_path = "C:/Users/dubey/OneDrive/Desktop/Assignment from IDZ/IDZ Digital (1).csv"
data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()

# Create a Pandas Excel writer using XlsxWriter as the engine.
excel_path = "report.xlsx"
writer = pd.ExcelWriter(excel_path, engine='xlsxwriter')

# Write the DataFrame data to the Excel file.
data.to_excel(writer, sheet_name='Data', index=False)

# Get the xlsxwriter workbook and worksheet objects.
workbook = writer.book
worksheet = writer.sheets['Data']

# Set the width of the columns for better readability.
for i, col in enumerate(data.columns):
    column_len = data[col].astype(str).str.len().max()
    column_len = max(column_len, len(col)) + 2
    worksheet.set_column(i, i, column_len)

# Generate and save 20 different plots
plot_titles = [
    'Marital Status by Gender', 'Salary Distribution by Occupation', 'Age Distribution', 
    'Occupation Counts', 'Salary by Gender', 'Age by Gender', 
    'Marital Status Counts', 'Occupation by Marital Status', 'Salary by Occupation and Gender',
    'Age Distribution by Marital Status', 'Gender Distribution by Occupation', 'Top 5 Occupations by Salary',
    'Bottom 5 Occupations by Salary', 'Average Salary by Marital Status', 'Average Age by Occupation',
    'Salary by Age Group', 'Gender Ratio by Occupation', 'Occupation Salary Boxplot', 
    'Age Boxplot by Gender', 'Salary Boxplot by Gender', 'Marital Status by Occupation'
]

for i, title in enumerate(plot_titles):
    plt.figure(figsize=(10, 6))
    
    if 'Salary' in title:
        sns.histplot(data['Salary'], kde=True)
    elif 'Age' in title:
        sns.histplot(data['Age'], kde=True)
    elif 'Marital Status' in title:
        sns.countplot(x='Marital Status', hue='Gender', data=data)
    elif 'Gender' in title:
        sns.countplot(x='Gender', data=data)
    else:
        sns.countplot(x='Occupation', data=data)
    
    plt.title(title)
    plot_filename = f'plot{i+1}.png'
    plt.savefig(plot_filename)
    plt.close()

    # Insert each plot into the Excel file
    row = 25 + (i // 2) * 20
    col = 'A' if i % 2 == 0 else 'M'
    worksheet.insert_image(f'{col}{row}', plot_filename)

# Save and close the Excel file
writer.close()
```
